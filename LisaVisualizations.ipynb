```python
''''''
def count_responses(dict_key, df, N1 = True):
    count1 = count2 = count3 = count4 = count5 = countnan=0
    if N1 == True:
        for column_name in Column_dict_N1[dict_key]:
            for response in df[column_name]:
                if response == '1':
                    count1 +=1
                elif response == '2':
                    count2 +=1
                elif response == '3':
                    count3 += 1
                elif response == '4':
                    count4 += 1
                elif response == '5':
                    count5 += 1
                else:
                    countnan += 1
        return [count1, count2, count3, count4, count5]
    else:
        for column_name in Column_dict_N2[dict_key]:
            for response in df[column_name]:
                if response == '1':
                    count1 +=1
                elif response == '2':
                    count2 +=1
                elif response == '3':
                    count3 += 1
                elif response == '4':
                    count4 += 1
                elif response == '5':
                    count5 += 1
                else:
                    countnan += 1  
        return [count1, count2, count3, count4, count5]
```
