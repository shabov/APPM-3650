```python
import numpy as np
import matplotlib.pyplot as plt

# this tells Jupyter to embed matplotlib plots in the notebook
%matplotlib inline
```

```python
def sinc(x):
    return np.sin(x) / x

def d_sinc(x):
    "derivative of sinc-function"
    return np.cos(x)/x - np.sin(x)/x**2
```

```python
# evaluate functions at 1000 points evenly spaced in [-15, 15]
x = np.linspace(-15, 15, 1000)
f = sinc(x)
df = d_sinc(x)
```

```python
# plot the sinc-function and its derivative
fig, ax = plt.subplots()
ax.plot(x, f, color="red", label=r"$sinc(x)$")
ax.plot(x, df, color="blue", ls="--", label=r"$\frac{d(sinc(x))}{dx}$")


ax.set_title("Example Notebook Plot")
ax.set_xlabel(r"$x$ [radians]")

ax.grid(True)
ax.legend();
```

```python
def even_odd_sums(nums):
    sum_odd=0
    sum_even=0
    for n in nums:
        if n%2 !=0:
            sum_odd+=n
        elif n%2==0:
            sum_even+=n
    return (sum_even,sum_odd)
```

```python
even_odd_sums([-5,18,12,-3,-1])
```

```python
def first_last(words):
    add_first=[]
    for n in words:
        add_first.append(n+n[0])
    return add_first
```

```python
def str4(words):
    if len(words)<4:
        words+= ((4-len(words))*'x')
    return words[0:4]
    
```

```python
def plurals(nouns):
    nouns_plural=[]
    for n in nouns:
        if n[-1]=='s' or n[-2:]=='ch':
            nouns_plural.append(n+'es')
        else:
            nouns_plural.append(n+'s')
    return nouns_plural
```

```python
plurals(['boat','iris','bench'])
```

```python
def middle(nums):
    if len(nums)%2 !=0:
        index=len(nums)//2
        return nums[index]
    else:
        index=len(nums)//2
        return (nums[index-1]+nums[index])/2
```

```python
middle([4,-3,80,6,7.2]), middle([4,-3,80,81,6,7.2])
```
